```python
import pandas as pd 
import matplotlib.pyplot as plt 
from geopy.geocoders import Nominatim
import pydeck
```

```python
dataset = pd.read_excel("dataset.xlsx", skiprows=1)
```

```python
lista_valori = []
#count_df = pd.DataFrame()
for e in dataset[dataset["City of Research organization"].notna()]["City of Research organization"].str.split(";"):
    e = [el.strip() for el in e]
    e = list(set(e))
    for s in e:
        s = s.strip()
        lista_valori.append(s)
new_naples_df = pd.DataFrame(lista_valori,columns=["a"])
city_counts = new_naples_df.value_counts().to_dict()
```

```python
cities = list(city_counts.keys())
counts = list(city_counts.values())
cities = [city[0] for city in cities]
```

```python
# Initialize a list to store the coordinates of each city
coords_list = []

# Initialize the geolocator
geolocator = Nominatim(user_agent="my_app_name")

# Loop through each city in the dictionary and get its coordinates
for i,city in enumerate(cities[199:]):
    print(i)
    location = geolocator.geocode(city)
    if location:
        lat = location.latitude
        lon = location.longitude
        coords_list.append((city, lat, lon,counts[i]))
```

```python
200
```

```python
cities_df = pd.DataFrame(coords_list, columns=["City", "Latitude", "Longitude","Count"])
cities_df.to_csv("test_cities_prova.csv",index=False)
```

```python
layer = pydeck.Layer(
    'HexagonLayer',
    cities_df,
    get_position=['Longitude', 'Latitude'],
    auto_highlight=True,
    elevation_scale='Count',
    pickable=True,
    elevation_range=[min(counts), max(counts)],
    extruded=True,
    coverage=1
)
```

```python
# Set the viewport location
view_state = pydeck.ViewState(
    longitude=-1.415,
    latitude=52.2323,
    zoom=6,
    min_zoom=5,
    max_zoom=15,
    pitch=40.5,
    bearing=-27.36)

# Combined all of it and render a viewport
r = pydeck.Deck(layers=[layer], initial_view_state=view_state)
r.to_html('hexagon-example.html')
```

```python
import pandas as pd
import folium

# creo il dataframe
df = pd.DataFrame({
    'City': ['Naples', 'Rome', 'Milan', 'Turin'],
    'Latitude': [40.835885, 41.893320, 45.464194, 45.067755],
    'Longitude': [14.248768, 12.482932, 9.189635, 7.682489],
    'Count': [2964, 965, 775, 471]
})

# trovo il count massimo tra le città
max_count = df['Count'].max()

# creo la mappa
m = folium.Map(location=[41.9028, 12.4964], zoom_start=6)

# aggiungo i marker per ogni città
for i, row in df.iterrows():
    # imposto lo stile del marker in base al count della città
    if row['Count'] == max_count:
        color = 'red'
        fill_color = 'crimson'
    else:
        color = 'gray'
        fill_color = 'black'

    # aggiungo il marker sulla mappa
    folium.CircleMarker(location=[row['Latitude'], row['Longitude']],
                        radius=row['Count'] / 100,  # dimensione del marker in base al count
                        color=color,
                        fill=True,
                        fill_color=fill_color,
                        fill_opacity=0.7,
                        tooltip=f"{row['City']}: {row['Count']}"
                        ).add_to(m)

# visualizzo la mappa
m
```
